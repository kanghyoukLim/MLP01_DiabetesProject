#### 전처리
데이터셋 일부 칼럼 값이 0이고 변수 범위가 달라 모델 성능 분석에 어려움이 있다. 전처리를 통해 개선해보자

결측값이 있는지 확인

```python
print(df.isnull().any())
```

```python
# 결측값이 없다

# 데이터셋 통계 요약 출력
print(df.describe())
```

min 값 0은 Pregnancies 는 가능하지만 다른 칼럼에서는 0이 될 수 없다. 결측값이다.  
각 칼럼에 0이 몇 개있는지 확인

```python
print("각 변수에 0인 row 수")
for col in df.columns:
    missing_rows = df.loc[df[col]==0].shape[0]
    print(col + ":" + str(missing_rows))
```

Insulin 칼럼에 0이 374개로 매우 많다. 전체 768개중 절반수준이다.  
따라서 이 값을 버리면 모델 성능이 나빠질 수 있다.  
  
  결측값처리방법
  - 결측값이 있는 로우제거(즉 해당하는 사람의 다른 데이터도 함께 제거)
  - 정상값의 평균이나 중앙값, 또는 최빈값으로 결측값을 대신한다. 
  - 머신 러닝 모델을 사용해 결측값을 예측한다.  
  
  Glucose등 연속 변수이므로 정상값의 평균을 사용해서 결측값을 바꾸자

```python
# 0을 NaN으로 바꿔서 판다스가 결측값을 인식하게 한다.

import numpy as np
df['Glucose']=df['Glucose'].replace(0,np.nan)
df['BloodPressure']=df['BloodPressure'].replace(0,np.nan)
df['SkinThickness']=df['SkinThickness'].replace(0,np.nan)
df['Insulin']=df['Insulin'].replace(0,np.nan)
df['BMI']=df['BMI'].replace(0,np.nan)
```

```python
print("각 변수에 0인 row 수")
for col in df.columns:
    missing_rows = df.loc[df[col]==0].shape[0]
    print(col + ":" + str(missing_rows))
```

```python
# NaN을 정상 값의 평균으로 변경. 판다스의 fillna() 함수 사용
df['Glucose']=df['Glucose'].fillna(df['Glucose'].mean())
df['BloodPressure']=df['BloodPressure'].fillna(df['BloodPressure'].mean())
df['SkinThickness']=df['SkinThickness'].fillna(df['SkinThickness'].mean())
df['Insulin']=df['Insulin'].fillna(df['Insulin'].mean())
df['BMI']=df['BMI'].fillna(df['BMI'].mean())
```

### 데이터표준화
데이터 표준화는 데이터셋의 각 숫자 변수를 평균이 0, 분산을 1로 변화하는 것  
신경망 역전파 알고리즘이 잘 동작하려면 표준화 필요, 변수가 큰 값에 더 많은 영향을 받을 수 있기때문  

```python
from sklearn import preprocessing
df_scaled = preprocessing.scale(df)
 
# preprocessig.scale() 함수는 판다스 DataFrame이 아닌 다른 객체를 반환하므로 DataFrame으로 다시 변환
df_scaled = pd.DataFrame(df_scaled, columns=df.columns)

#Outcome 칼럼은 목표변수이므로 표준화 필요 없다. 그대로 사용.
df_scaled['Outcome'] = df['Outcome']
df = df_scaled

#  변환한 데이터셋에서 각 변수의 평균, 표준편차, 최댓값을 계산하여 출력
print(df.describe().loc[['mean', 'std', 'max'],].round(2).abs())
```

### 데이터셋 분할
- 훈련 데이터셋
- 검증 데이터셋
- 테스트 데이터셋

검증 데이터셋은 훈련을 조기 종료할지 결정에 사용. 즉 예측 성능이 더 이상 좋아지지 않는 시점에 훈련을 종료한다는 것이며 이는 과적합을 방지하는 효과가 있다.  
테스트 데이터셋은 신경망 훈련에 사용되지 않는다. 모델 성능을 최종 평가하는 데 사용된다. 이를 통해 모델의 실 성능을 비교적 정확하게 반영할 수 있다. 

```python
from sklearn.model_selection import train_test_split

# 데이터에서 독립변수와 종속변수 지정
X = df.loc[:, df.columns != 'Outcome']
y = df.loc[:, 'Outcome']

# 훈련과 테스트 데이터셋 나눔(8:2)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# 훈련과 검증 데이터셋 나눔 <-- 확인 필요
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2)

```

### 다중 레이어 퍼셉트론  
EDA(Exploratory Data Analysis, 탐색적 데이타 분석)과 데이터 전처리 완료  
신경망 아키텍처 설계, 각 레이어에 비선형 활성화 함수 세팅

모델 이키텍처 : 입력레이어 - 은닉레이어 - 은닉레이어 - 출력레이어  
입력 레이어의 각 노드는 데이터셋의 각 특징 변수(즉, 칼럽)와 같다. 피마 인디안 데이터셋에는 독립변수가 8개 있으므로 입력 레이어의 노드도 8개 이다.  
은닉 레이어 
활성화 함수 : 렐루(ReLu)와 시그모이드(sigmoid) 활성화 함수를 사용  

### 케라스 모델 만들기

```python
from keras.models import Sequential
from keras.layers import Dense

# Sequential 클래스 만듦
model = Sequential()

# 첫 번째 은닉레이어 추가(노드32는 임의로 설정)
model.add(Dense(32, activation='relu', input_dim=8))

# 두 번째 은닉레이어 추가
model.add(Dense(16, activation='relu'))


# 출력 레이어 추가(출력 레이어는 당뇨, 비당뇨를 결정하는 이진분류이므로 노드는 1개, 출력값을 0과 1사이로 나타내는 시그모이드 활성화 함수 사용)
model.add(Dense(1, activation='sigmoid'))
```

### 모델 컴파일
모델 훈련 전에 필요한 매개변수를 정의해야 함  
- 옵티마이저 : 아담
- 손실함수 : binary_crossentropy
- 평가지표(metrics) : accuracy - 올바르게 분류된 샘플의 비율


```python
# 모델 컴파일
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# 모델 훈련
model.fit(X_train, y_train, epochs=200)
```

### 결과분석
다중 레이어 퍼셉트론을 훈련시켰다.   
테스트 정확도, 혼동 행력, ROC곡선을 사용해 모델을 최종 평가한다. 

```python
# 테스트 정확도
#모델 성능은 evaluate()함수로 평가한다. 훈련 데이터셋과 테스트 데이터셋을 모두 사용해보자.

scores = model.evaluate(X_train, y_train, verbose=False)
print("Training Accuracy: %.2f%%\n" % (scores[1]*100))

scores = model.evaluate(X_test, y_test, verbose=False)
print("Testing Accuracy: %.2f%%\n" % (scores[1]*100))
```

훈련 데이터셋으로 평가한 Training accuracy는 100%, 훈련에 사용되지 않은 Test 데이터를 이용한 평가는 70.13%의 정확도가 나왔다.  

```python
scores = model.evaluate(X_val, y_val, verbose=False)
print("Testing Accuracy: %.2f%%\n" % (scores[1]*100))
```

val 데이터셋도 훈련에 사용되지 않는다. 다만 향후 데이터 증량할 때는 주의해야 한다.

#### 혼동 행렬(Confusion matrix)
사이킷런 패키지에 포함되어 있으면 seaborn으로 시각화 할 수 있다.

```python
y_test_pred_prob = model.predict(X_test)
y_test_pred = (y_test_pred_prob > 0.5).astype(int)

c_matrix = confusion_matrix(y_test, y_test_pred)
ax = sns.heatmap(c_matrix, annot=True, xticklabels=['No Diabetes', 'Diabetes'], yticklabels=['No Diabetes', 'Diabetes'], cbar=False, cmap='Blues')
ax.set_xlabel("Prediction")
ax.set_ylabel("Actual")
plt.show()
plt.clf()
```

Prediction으로 Positive와 Negative 구분, 예측과 실제가 맞으면 True, 예측과 실제가 다르면 False  
True Positive 는 34건, True Negative 는 81건, False Positive 는 17건, False Negative 는 22건  
당뇨 예측에서 False Negative 는 False Positive 보다 더 위함할 수 있다. 따라서 여러 예측 모델이 있다면 False negative가 낮은 모델을 선택해야 한다.

#### ROC곡선
ROC 곡선은 FPR, False Positive Rate를 X축에 놓고 TPR, True Positive Rate를 Y축에 그린 차트다.  
FPR은 실제가 Negative의 전체예측 중 FP값의 비율 즉, FP/(TN+FP), 이 의미는 참이 아닌 것을 참이라고 잘 못 예측한 비율  
TPR은 실제가 Positive의 전체예측 중 TP값의 비율 즉, TP/(TP+FN), 이 의미는 참인 것을 참이라고 제대로 예측한 비율  
ROC곡선으로 모델 성능을 평가하는 척도는 AUC(Area Under the Curve)다.  
모델의 AUC가 크다면 클래스 분류 정확도가 높다는 뜻이며, AUC가 낮다면 모델 정확도가 나쁘다는 것이다.  
곡선이 차트의 대각선에 가까운 모델은 무작위로 분류하는 것보다 나을 게 없다는 것이다.

```python
# 당뇨 예측 모델의 ROC곡선을 그리고 성능 평가
from sklearn.metrics import roc_curve

import matplotlib.pyplot as plt
```

```python
y_test_pred_probs = model.predict(X_test)
FPR, TPR, _ = roc_curve(y_test, y_test_pred_probs)
plt.plot(FPR, TPR)
plt.plot([0,1],[0,1],'--', color='black') # 대각선
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
plt.clf()
```
